## Removing Outliers

```python
sns.boxplot(x=df["total_sqft"])
plt.show()
```

```python
from feature_engine.outliers import Winsorizer
win=Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables=("total_sqft"))
df_win=win.fit_transform(df[['total_sqft']])
df["total_sqft"]=df_win
print(win.left_tail_caps_,win.right_tail_caps_)
```

```python
sns.boxplot(x=df["total_sqft"])
plt.show()
```

```python
sns.boxplot(x=df["bath"])
plt.show()
```

```python
win=Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables=("bath"))
df_win=win.fit_transform(df[['bath']])
df["bath"]=df_win
print(win.left_tail_caps_,win.right_tail_caps_)
```

```python
sns.boxplot(x=df["bath"])
plt.show()
```

```python
sns.boxplot(x=df["bhk"])
plt.show()
```

```python
win=Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables=("bhk"))
df_win=win.fit_transform(df[['bhk']])
df["bhk"]=df_win
print(win.left_tail_caps_,win.right_tail_caps_)
```

```python
sns.boxplot(x=df["bhk"])
plt.show()
```

```python
sns.boxplot(x=df["price"])
plt.show()
```

```python
win=Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables=("price"))
df_win=win.fit_transform(df[['price']])
df["price"]=df_win
print(win.left_tail_caps_,win.right_tail_caps_)
```

```python
sns.boxplot(x=df["price"])
plt.show()
```

```python
df.describe()
```

```python
X=df.drop(columns='price')
y=df['price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
model=LinearRegression()
model.fit(X_train,y_train)
train_pred=model.predict(X_train)
test_pred=model.predict(X_test)
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
score=cross_val_score(model,X,y,cv=5)
print(score)
cv_score=score.mean()
print("Best Score::",cv_score)
```
